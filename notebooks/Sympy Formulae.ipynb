```python
import sympy as sp
```

```python

def NF(sigma_max, sigma_min, UTS, kd, kg):
    delta_Z = (sigma_max - sigma_min) / (1 - (sigma_min+sigma_max)/(2*UTS))**2
    return 5622*(UTS/(delta_Z*kd*kg))**5.26

sigma_max_1, sigma_min_1, UTS, kd, kg = sp.symbols('sigma_max_1, sigma_min_1, UTS, kd, kg')
sigma_max_2, sigma_min_2 = sp.symbols('sigma_max_2, sigma_min_2')
N1, N2 = sp.symbols('N1, N2')

NF1 = sp.Eq(NF(sigma_max_1, sigma_min_1, UTS, kd, kg),N1)
NF2 = sp.Eq(NF(sigma_max_2, sigma_min_2, UTS, kd, kg),N2)
NF1 = NF1.subs(sigma_min_1,0)
NF2 = NF2.subs(sigma_min_2,0)

N2_solved = sp.solve(sp.Eq(NF2.lhs/NF1.lhs, NF2.rhs/NF1.rhs),'N2')
# display(sp.simplify(NF2/NF1))
```
