```python
from SALib.sample import saltelli
from SALib.analyze import sobol

def pof_type(F_CR,FA,W,WA,FD,MF):
    return ( 1-( (1-F_CR/10)*(1-FA/10)*(1-W/10)*(1-WA/10)*(1-FD/10) ) )*MF

problem = {
    'num_vars':6,
    'names':['F_CR','FA','W','WA','FD','MF'],
    'bounds':[[0, 10],
             [0, 10],
             [0, 10],
             [0, 10],
             [0, 10],
             [0, 10]]
}

param_values = saltelli.sample(problem,1000)

Y = np.zeros([param_values.shape[0]])

for i, X in enumerate(param_values):
    Y[i] = pof_type(*X)
```

```python
Si = sobol.analyze(problem, Y)
```

```python
sns.barplot(y='names', x='sensitivities', data=pd.DataFrame(dict(names = problem['names'], sensitivities=Si['ST'])).sort_values('sensitivities',ascending=False))
plt.xscale('log')
# Si['ST']
```

```python
param_values.mean(axis=0)
```

```python
from SALib.sample import morris as ms
from SALib.analyze import morris as ma
from SALib.plotting import morris as mp

morris_problem = {
    'num_vars':6,
    'names':['F_CR','FA','W','WA','FD','MF'],
    'bounds':[[0, 10],
             [0, 10],
             [0, 10],
             [0, 10],
             [0, 10],
             [0, 10]]
}

number_of_trajectories = 1000

sample = ms.sample(morris_problem, number_of_trajectories, num_levels=4)

output = pof_type(*sample.T)
Si = ma.analyze(morris_problem, sample, output, print_to_console=False)
print("{:20s} {:>7s} {:>7s} {:>7s}".format("Name", "mu", "mu_star", "sigma"))
for name, s1, st, mean in zip(morris_problem['names'], 
                              Si['mu'], 
                              Si['mu_star'], 
                              Si['sigma']):
    print("{:20s} {:=7.2f} {:=7.2f} {:=7.2f}".format(name, s1, st, mean))
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
mp.horizontal_bar_plot(ax1, Si, {})
mp.covariance_plot(ax2, Si, {})
ax1.tick_params(axis='y', labelsize=22)
```

# Boxplot Generation

```python
for x in df.select_dtypes(np.float64).iteritems():#.describe()
    plt.figure(figsize=(10,5))
    if x[1].value_counts().empty:
        pass
    else:
        if x[1].pipe(np.log10).replace([np.inf,-np.inf], np.nan).pipe(np.nanmin) < -2.0:
            plt.yscale('log')
            plt.ylim(x[1].replace(-np.inf, np.nan).dropna().pipe(np.nanmin)/1.1,x[1].replace(np.inf, np.nan).dropna().pipe(np.nanmax)*1.1)
            plt.ylim(1e-10,10.0)
        else:
            plt.yscale('linear')
            plt.ylim(0,x[1].replace([np.inf,-np.inf], 0.0).pipe(np.nanmax)*1.1)
        x[1].plot.box(sym='k.')
        plt.savefig(x[0].replace('/','_').replace('.','_')+'.png', bbox_inches='tight')
    

# df.select_dtypes(np.float64).plot(kind='box',subplots=True, layout=(df.select_dtypes(np.float64).shape[1]//2,2))
```
